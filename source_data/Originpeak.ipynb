The narrowPeak datasets used in this study are available at:
https://zenodo.org/records/20011733

Below is a pipeline to merge peak files and extract corresponding genome sequences.

# Step 1: Find all .bed and .narrowPeak files (excluding merged.bed)

```python
! find . -type f \( -name "*.bed" -o -name "*.narrowPeak" \) \
! -name "merged.bed" \
| sort \
| xargs cat \
| awk '{print $1,$2,$3}' OFS="\t" \
| sort -k1,1 -k2,2n \
| uniq > originPeak.bed
```

# Step 2: Download reference genome (hg19) from UCSC

```python
! wget https://hgdownload.soe.ucsc.edu/goldenPath/hg19/bigZips/hg19.fa.gz
```

# Step 3: Decompress the genome file

```python
! gunzip hg19.fa.gz
```

# Step 4: Extract DNA sequences for peak regions using bedtools
-fi: input fasta (reference genome)
-bed: peak regions
-fo: output fasta file

```python
! bedtools getfasta -fi hg19.fa -bed originPeak.bed -fo originPeak.fa
```
